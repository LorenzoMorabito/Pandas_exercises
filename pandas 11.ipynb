### Write a Pandas program to split the following given dataframe into groups based on school code and class.

```python
student_data.groupby('school_code').agg({
    'class': 'count',
    'name': 'count',
    'age': 'mean',
    'height': ['min','max', lambda x: int(x.mean())],
    'weight': ['min','max','mean'],
    'address': ['count', 'nunique']
}).reset_index().rename(columns= {'<lambda_0>': 'mean'})
```
