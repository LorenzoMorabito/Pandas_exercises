### Create a dataframe of ten rows, four columns with random values. Convert some values to nan values. Write a Pandas program which will highlight the nan values.

```python
import numpy as np 
import pandas as pd
```

```python
# Creo un df con size(10,4)
df12 = pd.DataFrame( np.random.rand(10,4))
df12
```

```python
# Creo due array casuali i cui valori user√≤ per generare le posizioni dei NaN
arr_x = np.random.randint(low=0, high=4, size=10) # indice asse x
arr_y = np.random.randint(0,9,10)                 # indice y

i = 0
while i <= arr_x.size-1:
    df12.at[arr_y[i], arr_x[i]] = np.nan          # df.at[row. column] -> ibrido df.loc/df.iloc
    i += 1
df12
```

```python
df12.style.applymap(lambda x: 'background-color: yellow' if np.isnan(x) else '' )
# .style      -> attributo che permette di customizzare i df 
# .applymap() -> metodo che permette di applicare una funzione a tutto il df (piu lento di apply)
# 'background-color: yellow' -> customizzazione da applicare 
# if np.isnan(x)             -> operatore condizionale -> .isnan(x) -> funzione che restituisce un booleano
# else ''                    -> operatore condizionale -> '' -> rappresentano nessuno stile
                                                    #  -> 'background-color: yellow'
```
